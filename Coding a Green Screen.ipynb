```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2
import os
from IPython import display
```

```python
IMAGE_PATH = "images"
```

This is the image we will be using in this notebook

```python
IMAGE = "1.jpg"
display.Image(os.path.join(IMAGE_PATH,IMAGE))
```

And this is the background we'll be using

```python
BACKGROUND = "space_1.jpg"
display.Image(os.path.join(IMAGE_PATH,BACKGROUND))
```

```python
image = cv2.imread(os.path.join(IMAGE_PATH,IMAGE))#Read the image
image = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)#convert image from BGR to RGB
plt.imshow(image)#Show image
```

```python
%matplotlib inline
plt.imshow(image)
```

```python
background = cv2.imread(os.path.join(IMAGE_PATH,BACKGROUND))#Read the image
background = cv2.cvtColor(background , cv2.COLOR_BGR2RGB)#convert image from BGR to RGB
plt.imshow(background)#Show Image
```

```python
print("Image Dimension:",image.shape)
print("Background Dimension:",background.shape)
```

```python
#Setting the Upper and Lower Boundary
lower_bound = np.array([0,55,0])
upper_bound = np.array([97,255,180])
```

```python
%matplotlib inline
mask = cv2.inRange(image,lower_bound,upper_bound)
plt.imshow(mask,cmap="gray")
```

```python
image[mask != 0] = 0
plt.imshow(image)
```

```python
height = mask.shape[0]
width = mask.shape[1]
```

```python
background_resize = cv2.resize(background,(width,height))
```

```python
background_resize[mask == 0] = 0
plt.imshow(background_resize)
```

```python
complete_image = background_resize + image
plt.imshow(complete_image)
```
